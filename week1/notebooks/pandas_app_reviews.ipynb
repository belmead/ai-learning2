# Selection

## Getitem
For a DataFrame, passing a single label selects a column and yields a Series equivalent to `df.A`.

```python
df["title"]
```

## Getitem
For a DataFrame, passing a slice `:` selects matching rows.

Can slice using `str`s as well (`df["20130102":"20130104"]`)

```python
df[0:3]
```

## Selection by label

Access a group of rows and columns by label(s) or a boolean array with `df.loc`.

`.loc[]` is primarily label based, but may also be used with a boolean array.

Selecting a row matching a label:

```python
df.loc[0]
```

Selecting all rows (`:`) with select column labels:

```python
df.loc[:, ['title', 'rating']]
```

For label slicing, both endpoints are included:

```python
df.loc[0:3, ['title']]
```

## Selection by position

`df.iloc` for purely integer-location based indexing for selection by position.

Select via the position of the passed integers:

```python
df.iloc[3]
```

Integer slices acts similar to NumPy/Python:

```python
df.iloc[0:3]
```

Lists of integer position locations.

Can be used to index both axes (seen below):

```python
df.iloc[[1, 2, 4], [0, 2]]
```

For slicing rows explicitly:

```python
df.iloc[1:3, :]
```

For slicing columns explicitly:

```python
df.iloc[:, 1:3]
```

For getting a value explicitly:

```python
df.iloc[1, 1]
```

For getting fast access to a scalar (equivalent to the prior method):

```python
df.iat[1, 1]
```

## Boolean Indexing

Select rows where `df.ratings` is greater than or equal to `4`.

```python
df[df["rating"] >= 4]
```

Selecting values from a DataFrame where a boolean condition is met:

```python
df[df["rating"] > 0]
```

Using `isin()` method for filtering:

```python
df2 = df.copy() # this function copies a DataFrame
df2["title"] = ["testTitle1", "testTitle2", "testTitle3", "testTitle4"] # note: this doesn't work, because this is an array of 4 values, and the index length is 500.
                                                                        # instead, using df2["title"] = "testTitle1" would work, since it's a scalar!

df2
df2[df2["title"].isin(["testTitle1", "testTitle3"])]
```

# Setting

Setting a new column automatically aligns the data by the indexes:

```python
s1 = pd.Series([1,2,3,4,5,6], index=pd.date_range("20130102", periods=6))
s1
df["F"] = s1
```

Setting values by label:

```python
df.at["title"[0], "A"] = 0
```

Setting values by position:

```python
df.iat[0, 1] = 0
```

Setting by assigning with a NumPy array:

```python
df.loc[: ,"D"] = np.array([5] * len(df))
```

The result of the prior setting operations:

```python
df
```

A `where` operation with setting:

```python
df2 = df.copy()
df2[df2 > 0] = -df2 # does not work here, but would inverse any numbers (ints, flotats)
df2
```
